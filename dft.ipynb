```python
import numpy as np
import matplotlib.pyplot as plt

N = 60
x = np.linspace(-2,10,N)
y = np.exp(-x**2)

plt.plot(x,y)
plt.show()
```

```python
c = np.zeros(N,complex)
n = np.arange(N)
for k in range(N):
  c[k] = np.sum(y*np.exp(-2j*np.pi*k*n/N))
plt.plot(np.real(c))
plt.plot(np.imag(c))
plt.plot(abs(c))
plt.show()
```

```python
c1 = np.zeros(N,complex)
k = np.arange(N)
for n in range(N):
  c1[n] = (1/N)*np.sum(c*np.exp(2j*np.pi*k*n/N))
plt.plot(np.real(c1))
#plt.plot(abs(c1))
plt.show()
```
